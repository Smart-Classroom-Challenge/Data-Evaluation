```python
#Auswertung Lüften
Auswertung_Delta = pd.DataFrame(columns=['delta_time', 'delta_co2', 'delta_temperature', 'delta_humidity'])
for klasse in Lüften:
    count_ventialtion_per_day = 0
    ind = 0
    for datapoint in Lüften[klasse]:
        #print(datapoint)
        if (ind >= 1): 
            #print("last")
            #print(Lüften[klasse][ind - 1])
            if (Lüften[klasse][ind - 1]['from'][0:10] != datapoint['from'][0:10]):
                count_ventialtion_per_day = 0    
        measuremen_from = l.get_measurement(klasse, startDate = datapoint['from'], endDate=datapoint['to'])
        
        
        
        max_measurement = measuremen_from['time'].max()
        min_measurement = measuremen_from['time'].min()
        delta_time = max_measurement - min_measurement
        #print(delta_time)
        #print(measuremen_from['co2'].max())
        #print(measuremen_from['co2'].min())
        #print("max - min")
        #print("equals")
        count_ventialtion_per_day = count_ventialtion_per_day + 1
        #print(count_ventialtion_per_day)
        #print(measuremen_from['co2'].max() - measuremen_from['co2'].min())
        delta_co2 = measuremen_from['co2'].max() - measuremen_from['co2'].min()
        delta_temperatur = measuremen_from['temperature'].max() - measuremen_from['temperature'].min()
        delta_humidity = measuremen_from['humidity'].max() - measuremen_from['humidity'].min()
        #print(klasse, datapoint['from'], datapoint['to'], delta_time, delta_co2, delta_temperatur, delta_humidity)
        #rint(min_measurement)
        #print({'delta_time': delta_time, 'delta_co2': delta_co2, 'delta_temperature': delta_temperatur, 'delta_humidity': delta_humidity})
        Auswertung_Delta= Auswertung_Delta.append({'co2_range_low': f"{measuremen_from['co2'].min()}",
                                                   'co2_range_high': f"{measuremen_from['co2'].max()}",
                                                   'temp_range_low': f"{measuremen_from['temperature'].min()}",
                                                   'temp_range_high': f"{measuremen_from['temperature'].max()}",
                                                   'hum_range_low': f"{measuremen_from['humidity'].min()}",
                                                   'hum_range_high': f"{measuremen_from['humidity'].max()}",
                                                   'Vent_cycle': f"{count_ventialtion_per_day}",
                                                   'delta_time': delta_time.seconds, 'delta_co2': delta_co2, 'delta_temperature': delta_temperatur, 'delta_humidity': delta_humidity}, ignore_index=True)
        ind = ind + 1

Auswertung_Delta = Auswertung_Delta.astype({'co2_range_low': np.float64})
Auswertung_Delta = Auswertung_Delta.astype({'co2_range_high': np.float64})
Auswertung_Delta = Auswertung_Delta.astype({'temp_range_low': np.float64})
Auswertung_Delta = Auswertung_Delta.astype({'temp_range_high': np.float64})
Auswertung_Delta = Auswertung_Delta.astype({'hum_range_low': np.float64})
Auswertung_Delta = Auswertung_Delta.astype({'hum_range_high': np.float64})
Auswertung_Delta = Auswertung_Delta.astype({'delta_time': np.float64})
Auswertung_Delta = Auswertung_Delta.astype({'Vent_cycle': np.float64})
Auswertung_Delta       
        
```
